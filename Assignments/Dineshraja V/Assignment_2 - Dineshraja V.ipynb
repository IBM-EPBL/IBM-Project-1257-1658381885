**6.Find the outliers and replace the outliers.**

**Finding the outliers using InterQuartile Range**

```python
Q1 = df['EstimatedSalary'].quantile(0.25)
Q3 = df['EstimatedSalary'].quantile(0.75)
IQR = Q3 - Q1
whisker_width = 0.5
Fare_outliers = df[(df['EstimatedSalary'] < Q1 - whisker_width*IQR) | (df['EstimatedSalary'] > Q3 + whisker_width*IQR)]
Fare_outliers.head()
```

**Replacing the outliers**

```python
Q1 = df['EstimatedSalary'].quantile(0.25)    
Q3 = df['EstimatedSalary'].quantile(0.75)
IQR = Q3 - Q1
whisker_width = 0.5
lower_whisker = Q1 - (whisker_width*IQR)
upper_whisker = Q3 + (whisker_width*IQR)
index=df['EstimatedSalary'][(df['EstimatedSalary']>upper_whisker)|(df['EstimatedSalary']<lower_whisker)].index
df.drop(index,inplace=True)
```
